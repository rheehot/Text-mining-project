```python
import zipfile
#html 태그 제거
import re
#> 제거
from bs4 import BeautifulSoup
from konlpy.tag import Hannanum, Kkma, Komoran, Okt

#파일명 추출
data_zip = zipfile.ZipFile('data (2).zip')
dataList = data_zip.namelist()
```

```python
dataList
```

```python
#데이터 불러오는 함수 만들기
def tokenizer(list):
    
    textFile = []
    for i in range(len(list)):
        
        f = open('data (2)/statement{}.txt'.format(i),'r',encoding='utf-8')
        line = f.readlines()
        
        
        #태그 제거
        textList = []
        pattern = '<[^>]*>|\n'
    
        line = re.sub(pattern=pattern, repl='', string=line[0])
        line = BeautifulSoup(line, 'html.parser').get_text()
        line = line.replace("[", "").replace("<", "").replace(">","").replace("]", "")
        
        textList.append(line)
        #print(textList)
        
        #처리한 데이터를 textFile 리스트에 저장
        textFile.append(textList)
        print(textFile)
    
    return textFile
```

```python
tokens = tokenizer(dataList)
```

```python
kkma= Kkma()
print(kkma.morphs(tokens[0][0]))
```

```python
def kkma_pos(tokens_list):
    kkma= Kkma()
    kkma_pos = []
    
    for i in tokens_list:
        #print(i[0])
        x = kkma.pos(i[0])
        #print(x)
        kkma_pos.append(x)
    return kkma_pos
```
